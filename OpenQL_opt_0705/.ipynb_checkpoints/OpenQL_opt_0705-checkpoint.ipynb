イジングハミルトニアン*とは？
$$
H= \sum_{\alpha} h_{\alpha} \bigotimes^{N}_{j=1}\sigma^{\alpha_j}_{j}\\
\alpha_j \in \{I, X , Y, Z\}
$$
Pauli termの重み付き和  
例えば$H= 0.5Z_1 \otimes I_2 \otimes I_3 + 0.3 X_1 \otimes Y_2 \otimes Z_3$ のようなもの。行列。  
$0.5ZIII$のように添字とテンソル積の記号を省略することもあります。  
$*$ 3次以上の項も含む

例として以下の問題を考える。
\begin{aligned}
   & \text{maximize}
       & \sum_{i} x_{i}\\
   & \text{subject to}
       & \sum_{i} i * x_{i}=3\\
       & & i \in \{1,2,3,4\} \\
       & & x_i \in \{0,1\}\\
\end{aligned}

```python
from docplex.mp.model import Model

from qiskit import BasicAer
from qiskit.aqua import Operator, run_algorithm
from qiskit.aqua.algorithms import VQE, ExactEigensolver
from qiskit.aqua.components.optimizers import SPSA
from qiskit.aqua.components.variational_forms import RY
from qiskit.aqua import QuantumInstance
from qiskit.aqua.translators.ising import docplex

# setup aqua logging
import logging
from qiskit.aqua import set_qiskit_aqua_logging
# set_qiskit_aqua_logging(logging.DEBUG)  # choose INFO, DEBUG to see the log
```

```python
# Create an instance of a model and variables
mdl = Model(name='max_vars')
x = {i: mdl.binary_var(name='x_{0}'.format(i)) for i in range(1,5)}

# Object function
max_vars_func = mdl.sum(x[i] for i in range(1,5))
mdl.maximize(max_vars_func)

# Constrains
mdl.add_constraint(mdl.sum(i*x[i] for i in range(1,5)) == 3)

print(mdl.export_to_string())
```

```python
qubitOp, offset = docplex.get_qubitops(mdl)
qubitOp.paulis
# An Ising Hamiltonian after conversion
```

結果の見方  
配列zのTrueになっている箇所にZがあって、配列xのTrueになっている箇所にXがある。Falseの所はIがあると見る。
\begin{aligned}
   H=
       & -9.5ZIII -19.5IZII -29.5IIZI -39.5IIIZ +5.0ZZII\\
    &+ 7.5ZIZI + 10.0ZIIZ + 15.0IZZI + 20.0IZIZ + 30.0IIZZ
\end{aligned}

左の最適化問題は右のイジングハミルトニアンに変換できる。右のイジングハミルトニアンをVQEやQAOAの入力とすることで左の最適化問題をQiskit Aquaで解くことが可能

変換方法について
1. 元の問題を制約なしの問題に変換する
    - 制約をペナルティ項として目的関数の中に組み込む
2. 目的関数の中の変数$x_i$を$Z_i$で置き換える

![how_to_translate.png](how_to_translate.png)

1 元の問題を制約なしの問題に変換する
- 現状、制約をそのまま扱うことが出来ないので制約をペナルティ項として目的関数の中に組み込み制約なしの問題に変換する（Mは十分大きな数）
- ココロ: 制約を満たしていない解(Infeasibleな解)は大きな数になり最小値にならないようにしている

![to_unconstraint_model.png](to_unconstraint_model.png)

![infeasible.png](infeasible.png)

![infeasible2.png](infeasible2.png)

Unconstrained modelへの変換例
\begin{aligned}
   & \text{maximize}
       & \sum_{i} x_{i}\\
   & \text{subject to}
       & \sum_{i} i * x_{i}=3\\
       & & i \in \{1,2,3,4\} \\
       & & x_i \in \{0,1\}\\
\end{aligned}
上記の問題は下記のように変換する
\begin{aligned}
   & \text{maximize}
       & \sum_{i} x_{i} - M(3-\sum_{i} i * x_{i})^2\\
   & \text{subject to}
       &  i \in \{1,2,3,4\} \\
       & & x_i \in \{0,1\}\\
\end{aligned}

2 目的関数の中の変数$x_i$を$Z_i$で置き換える  
- $x_i\rightarrow (1-Z_i)/2$のように置き換える。$Z_i$の固有値は$\pm 1$なので$Z_i$が+1の時、$x_i$は0。$Z_i$が-1の時、$x_i$は1に対応
- $x_i$の否定は$1-x_i$なので$(1+Z_i)/2$で置き換える
- $x_i$はキュービット$i$に対応

![to_hamiltonian.png](to_hamiltonian.png)

$Z_i$での置き換えの具体例
\begin{aligned}
   & c_{i,j}x_i x_j \rightarrow
       & c_{i,j} \frac{(1-Z_i)}{2} \frac{(1-Z_j)}{2}\\
   & &= \frac{c_{i,j}}{4} (1-Z_i -Z_j +Z_i Z_j)
\end{aligned}
$IIZ_iIII$、$IIZ_iIZ_jI$のように$i,j$以外のキュービットはIで置き換える。  


これらの和が先に話したイジングハミルトニアン
\begin{aligned}
   H=
       & -9.5ZIII -19.5IZII -29.5IIZI -39.5IIIZ +5.0ZZII\\
    &+ 7.5ZIZI + 10.0ZIIZ + 15.0IZZI + 20.0IZIZ + 30.0IIZZ
\end{aligned}

このイジングハミルトニアンをVQEやQAOAの入力とすることで量子コンピューターで最適化問題が解ける

最適化問題からイジングハミルトニアンへの変換方法のまとめ
1. 元の問題を制約なしの問題に変換する
    - 制約をペナルティ項として目的関数の中に組み込む
    - ココロ: 制約を満たしていない解(Infeasibleな解)は大きな数になり最小値にならないようにしている
2. 目的関数の中の変数$x_i$を$Z_i$で置き換える
    - $x_i\rightarrow (1-Z_i)/2$のように置き換える。$Z_i$の固有値は$\pm 1$なので$Z_i$が+1の時、$x_i$は0。$Z_i$が-1の時、$x_i$は1に対応
    - $x_i$はキュービット$i$に対応

変換したイジングハミルトニアンをVQEやQAOAの入力として最適化問題を量子コンピューターで解く

最適化問題からイジングハミルトニアンへの変換方法とそのキモチは分かったでしょうか？  
ツールを使うだけなら必要ない知識かもしれませんが、これが分かっていれば自分で解きたい問題を変換してその問題を量子コンピューターで解けるようになります。ツールが存在しないような皆さんが実務で抱えている問題に応用でき、量子コンピューターでより面白いこと(変換方法の改善で論文を書いたり、Qiskitにコードが組み込まれたり etc...)が出来るようになると思います。それで量子コンピューターが面白いものと思えてもらえて魅力が伝われば幸いです。

## Problem-specific translators
- 名前の通り特定の問題専用のトランスレーター。例えばMaxcutやTSP(Traveling Salesman Problem, 巡回セールスマン問題)など
- それぞれのトランスレーターは特定の問題に対応している
- コードは`qiskit/aqua/translators/ising` フォルダーにある

### TSPトランスレーターの使用例

TSPとは
> 巡回セールスマン問題（じゅんかいセールスマンもんだい、英: traveling salesman problem、TSP）は、都市の集合と各2都市間の移動コスト（たとえば距離）が与えられたとき、全ての都市をちょうど一度ずつ巡り出発地に戻る巡回路のうちで総移動コストが最小のものを求める（セールスマンが所定の複数の都市を1回だけ巡回する場合の最短経路を求める）組合せ最適化問題である。

Wikipediaより

#### Preparation

`from qiskit.aqua.translators.ising import tsp`  
これがTSPのproblem-specific translators

```python
# Import packages 
import matplotlib.pyplot as plt
import matplotlib.axes as axes
%matplotlib inline
import numpy as np
import networkx as nx

from qiskit.aqua.input import EnergyInput
from qiskit.aqua import Operator, run_algorithm
from qiskit.aqua.algorithms import ExactEigensolver

from qiskit.aqua.translators.ising import tsp
```

```python
# Generating a graph of 3 nodes
n = 4
np.random.seed(100)
num_qubits = n ** 2
ins = tsp.random_tsp(n)
G = nx.Graph()
G.add_nodes_from(np.arange(0, n, 1))
colors = ['r' for node in G.nodes()]
pos = {k: v for k, v in enumerate(ins.coord)}
default_axes = plt.axes(frameon=True)
nx.draw_networkx(G, node_color=colors, node_size=600, alpha=.8, ax=default_axes, pos=pos)
print('distance\n', ins.w)
```

### TSPトランスレーターを用いて問題を対応するイジングハミルトニアンに変換

```python
qubitOp, offset = tsp.get_tsp_qubitops(ins)
algo_input = EnergyInput(qubitOp)
```

```python
def draw_tsp_solution(G, order, colors, pos):
    G2 = G.copy()
    n = len(order)
    for i in range(n):
        j = (i + 1) % n
        G2.add_edge(order[i], order[j])
    default_axes = plt.axes(frameon=True)
    nx.draw_networkx(G2, node_color=colors, node_size=600, alpha=.8, ax=default_axes, pos=pos)
```

```python
#Making the Hamiltonian in its full form and getting the lowest eigenvalue and eigenvector
ee = ExactEigensolver(qubitOp, k=1)
result = ee.run()

print('energy:', result['energy'])
print('tsp objective:', result['energy'] + offset)
x = tsp.sample_most_likely(result['eigvecs'][0])
print('feasible:', tsp.tsp_feasible(x))
z = tsp.get_tsp_solution(x)
print('solution:', z)
print('solution objective:', tsp.tsp_value(z, ins.w))
draw_tsp_solution(G, z, colors, pos)
```

### Maxcutトランスレーターの使用例

Maxcutとは  
グラフをの頂点集合を二分割にする（エッジで繋がっていない組に分ける）。その時のカットエッジの総数 (辺重みグラフの場合はカットエッジそれぞれの辺重みの総和) が最大なものを求める。  
以下のグラフは頂点数が5で、白と黒のように頂点を二分割するカット（赤い線がカットエッジ）が最大カット
<img src="./maxcut.png" width=30%>
画像はWikipediaより

#### Preparation

`from qiskit.aqua.translators.ising import max_cut`  
これがMaxcutのproblem-specific translators

```python
# Import packages 
import matplotlib.pyplot as plt
import matplotlib.axes as axes
%matplotlib inline
import numpy as np
import networkx as nx

from qiskit.aqua.input import EnergyInput
from qiskit.aqua import Operator, run_algorithm
from qiskit.aqua.algorithms import ExactEigensolver

from qiskit.aqua.translators.ising import max_cut
```

```python
# Generating a graph of 4 nodes 

n=4 # Number of nodes in graph
G=nx.Graph()
G.add_nodes_from(np.arange(0,n,1))
elist=[(0,1,1.0),(0,2,1.0),(0,3,1.0),(1,2,1.0),(2,3,1.0)]
# tuple is (i,j,weight) where (i,j) is the edge
G.add_weighted_edges_from(elist)

colors = ['r' for node in G.nodes()]
pos = nx.spring_layout(G)
default_axes = plt.axes(frameon=True)
nx.draw_networkx(G, node_color=colors, node_size=600, alpha=.8, ax=default_axes, pos=pos)
```

```python
# Computing the weight matrix from the random graph
w = np.zeros([n,n])
for i in range(n):
    for j in range(n):
        temp = G.get_edge_data(i,j,default=0)
        if temp != 0:
            w[i,j] = temp['weight'] 
print(w)
```

### Maxcutトランスレーターを用いて問題を対応するイジングハミルトニアンに変換

```python
qubitOp, offset = max_cut.get_max_cut_qubitops(w)
algo_input = EnergyInput(qubitOp)
```

```python
#Making the Hamiltonian in its full form and getting the lowest eigenvalue and eigenvector
ee = ExactEigensolver(qubitOp, k=1)
result = ee.run()

x = max_cut.sample_most_likely(result['eigvecs'][0])
print('energy:', result['energy'])
print('max-cut objective:', result['energy'] + offset)
print('solution:', max_cut.get_graph_solution(x))
print('solution objective:', max_cut.max_cut_value(x, w))

colors = ['r' if max_cut.get_graph_solution(x)[i] == 0 else 'b' for i in range(n)]
nx.draw_networkx(G, node_color=colors, node_size=600, alpha = .8, pos=pos)
```

## DOcplex translator
- より一般的 (特定の問題専用のトランスレーターではない)
- 最適化モデルから **自動** でイジングハミルトニアンを作成
    - ユーザーは最適化モデルを “DOcplex” で作成
        - DOcplexは最適化モデル用のPythonライブラリー
        - 最適化モデルはコードが短く、より直感的
        - Qiskit 0.9 (May, 2019) で追加された新機能

### DOcplexトランスレーターを用いてTSPを解く例
以下はDOcplexトランスレーターを用いて問題を解く際のステップ
1. 定式化した問題を最適化モデルをDOcplexを用いて書く
2. DOcplexトランスレーターを用いてモデルをイジングハミルトニアンに変換
3. 変換したイジングハミルトニアンを入力としVQEやQAOAを用いて問題を解く

以下が定式化したTSP
\begin{aligned}
   & \text{minimize}
       & \sum_{i,j}w_{ij}\sum_{p} x_{i,p}x_{j,p+1}& \\
   & \text{subject to}
       & \sum_{p} x_{i,p}=1 \\
       & &\sum_{i} x_{i,p}=1 \\
       & & x_{i,p} \in \{0,1\} & \\
       & & i \in \{1,...,n\} \\
       & & p \in \{1,...,n\}
\end{aligned}

```python
# Generating a graph of 3 nodes
n = 4
np.random.seed(100)
num_qubits = n ** 2
ins = tsp.random_tsp(n)
G = nx.Graph()
G.add_nodes_from(np.arange(0, n, 1))
colors = ['r' for node in G.nodes()]
pos = {k: v for k, v in enumerate(ins.coord)}
default_axes = plt.axes(frameon=True)
nx.draw_networkx(G, node_color=colors, node_size=600, alpha=.8, ax=default_axes, pos=pos)
print('distance\n', ins.w)
```

定式化したTSPを最適化モデルとして書く

```python
from docplex.mp.model import Model
from qiskit.aqua.translators.ising import docplex
n=4

# Create an instance of a model and variables
mdl = Model(name='tsp')
x = {(i,p): mdl.binary_var(name='x_{0}_{1}'.format(i,p)) for i in range(n) for p in range(n)}

# Object function
tsp_func = mdl.sum(ins.w[i,j] * x[(i,p)] * x[(j,(p+1)%n)] for i in range(n) for j in range(n) for p in range(n))
mdl.minimize(tsp_func)

# Constrains
for i in range(n):
    mdl.add_constraint(mdl.sum(x[(i,p)] for p in range(n)) == 1)
for p in range(n):
    mdl.add_constraint(mdl.sum(x[(i,p)] for i in range(n)) == 1)
```

DOcplexトランスレーターを用いてモデルからイジングハミルトニアンを作成

```python
qubitOp_docplex, offset_docplex = docplex.get_qubitops(mdl)
```

```python
#Making the Hamiltonian in its full form and getting the lowest eigenvalue and eigenvector
ee = ExactEigensolver(qubitOp_docplex, k=1)
result = ee.run()

print('energy:', result['energy'])
print('tsp objective:', result['energy'] + offset_docplex)
x = tsp.sample_most_likely(result['eigvecs'][0])
print('feasible:', tsp.tsp_feasible(x))
z = tsp.get_tsp_solution(x)
print('solution:', z)
print('solution objective:', tsp.tsp_value(z, ins.w))
draw_tsp_solution(G, z, colors, pos)
```

### DOcplexトランスレーターを用いてMaxcutを解く例

以下が定式化したMaxcut
\begin{aligned}
   & \text{maximize}
       & \sum_{i,j}w_{ij} x_i (1-x_j)& \\
   & \text{subject to}
       & x_{i} \in \{0,1\} & \\
       & & i \in \{1,...,n\} \\
       & & j \in \{1,...,n\}
\end{aligned}

```python
# Generating a graph of 4 nodes 

n=4 # Number of nodes in graph
G=nx.Graph()
G.add_nodes_from(np.arange(0,n,1))
elist=[(0,1,1.0),(0,2,1.0),(0,3,1.0),(1,2,1.0),(2,3,1.0)]
# tuple is (i,j,weight) where (i,j) is the edge
G.add_weighted_edges_from(elist)

colors = ['r' for node in G.nodes()]
pos = nx.spring_layout(G)
default_axes = plt.axes(frameon=True)
nx.draw_networkx(G, node_color=colors, node_size=600, alpha=.8, ax=default_axes, pos=pos)
```

```python
# Computing the weight matrix from the random graph
w = np.zeros([n,n])
for i in range(n):
    for j in range(n):
        temp = G.get_edge_data(i,j,default=0)
        if temp != 0:
            w[i,j] = temp['weight'] 
print(w)
```

```python
from docplex.mp.model import Model
from qiskit.aqua.translators.ising import docplex

# Create an instance of a model and variables.
mdl = Model(name='max_cut')
x = {i: mdl.binary_var(name='x_{0}'.format(i)) for i in range(n)}

# Object function
maxcut_func = mdl.sum(w[i,j]* x[i] * ( 1 - x[j] ) for i in range(n) for j in range(n))
mdl.maximize(maxcut_func)

# No constraints for MaxCut problems.
```

DOcplexトランスレーターを用いてモデルをイジングハミルトニアンに変換

```python
qubitOp_docplex, offset_docplex = docplex.get_qubitops(mdl)
```

```python
#Making the Hamiltonian in its full form and getting the lowest eigenvalue and eigenvector
ee = ExactEigensolver(qubitOp_docplex, k=1)
result = ee.run()

x = max_cut.sample_most_likely(result['eigvecs'][0])
print('energy:', result['energy'])
print('maxcut objective:', result['energy'] + offset_docplex)
print('solution:', max_cut.get_graph_solution(x))
print('solution objective:', max_cut.max_cut_value(x, w))

colors = ['r' if max_cut.get_graph_solution(x)[i] == 0 else 'b' for i in range(n)]
nx.draw_networkx(G, node_color=colors, node_size=600, alpha = .8, pos=pos)
```

### Docplexトランスレーターの現在の制約
現在は下記のモデルのみ変換可能
- バイナリ変数のみ 
- 目的関数の最大の次数は二次まで
- 等式制約のみ

入力したモデルは変換前にチェックされ、もしモデルがサポートしていない要素を含んでいる場合にはエラーが出力される

この制限のもとでも、いくつものtypicalな最適化問題を扱うことができる。
